```python
import json
def convert (json_string):
    obj = json.loads(json_string)
    l=[]
    for i in obj:
        l.append(i['name'])
    return l
```

```python
import json
def convert2 (json_string):
    obj = json.loads(json_string)
    l=[]
    count=0
    for i in obj:
        if count !=3:
            l.append(i['name'])
            count+=1
        else:
            break
            
        
    return l
```

```python
import json
def Director (json_string):
    obj = json.loads(json_string)
    l=[]
    for i in obj:
        if i['job']=='Director':
            l.append(i['name'])
            break
        
            
        
    return l
```

```python
movies['crew']=movies['crew'].apply(Director)
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return (" ".join(y))
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
def recommend(movie):
    # Check if the movie exists
    movie_indices = new_df[new_df['title'].str.lower() == movie.lower()].index
    if movie_indices.empty:
        return f"'{movie}' not found in the dataset."
    
    # Get movie index
    movie_index = movie_indices[0]

    # Find similar movies
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    # Return the recommended movie titles
    return [new_df.iloc[i[0]]['title'] for i in movie_list]

# Test the function

```
